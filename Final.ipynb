```python
import sys
import pylsl
import time

streams = pylsl.resolve_stream('type','signal')
inlet = pylsl.stream_inlet(streams[0])

sample = pylsl.vectorf()
while True:
    try:
        inlet.pull_sample(sample)
        sys.stdout.write(str(sample[0]) + ', ');
        time.sleep(0.1)
    except:
        time.sleep(0.1)
```
