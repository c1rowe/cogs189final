```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('chase_trial.csv')
df
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['Attention'])

start_i = df[df['Event Id'] == 33025.0].index.tolist()
found_i = df[df['Event Id'] == 33027.0].index.tolist()

for start in start_i:
    plt.axvline(x=start, color='green')
    
for found in found_i:
    plt.axvline(x=found, color='red')
    
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.psd(df['Low Alpha'], Fs=512)
plt.psd(df['Attention'], Fs=1)
plt.show()
```
