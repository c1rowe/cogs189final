```python
densities = df.iloc[0:(512*30),6:13].plot.kde(bw_method='scott', figsize=(14,8))
plt.xlim(-100, 75000)
plt.show()
```

```python
step = 100
wavemax = int(750000 / step)
```

```python
# matrix to contain all densities
# let rows == wave type, columns == pdf generated density Y at X amplitude (for given wave)
density_matrix = np.empty([7,int(wavemax)])

# iterate across all wave types and stack each wave type density as rows in density_matrix
# wave type density is a 75,000 value list. each value represents the density of waves at [index] frequency
for i in range(6,13):
    print(i)
    # potential optimization: clear repeating values in original df
    density = stats.gaussian_kde(df.iloc[0:(512*30),i], bw_method='scott')
    density_matrix[i-6] = list(density.pdf(range(0,750000,step)))     # potential optimization: increase step width
    
# now find the total average densities for all wave types, column wise (end with 1x75000 vector) that reprents overall behavior
overall_matrix = np.mean(density_matrix, axis=0)
```

```python
overall_matrix * 100
```

```python
# must create accurate x-axis vector
x_axis = list(range(0,wavemax))
x_axis = [x * step for x in x_axis]

plt.figure(figsize=(10,6))
plt.plot(x_axis, overall_matrix)
plt.ylabel('Density')
plt.legend(['Aggregate'])
plt.xlim(0, 15000)

densities = df.iloc[0:(512*30),6:13].plot.kde(bw_method='scott', figsize=(10,6))
plt.xlim(0, 15000)
plt.show()
```

```python
density_df = pd.DataFrame(density_matrix, index=df.iloc[1,6:13].index.tolist())
density_df = density_df.transpose()
density_df['Amplitude'] = (density_df.index + 1) * step
density_df.iloc[:1400,:]
```

```python
density_df.iloc[:,:-1].plot(figsize=(14,8))
plt.xlim(0,1000)
plt.show()
```
